Beautiful Soup 是一个可以从HTML或XML文件中提取数据的Python库.它能够通过你喜欢的转换器实现惯用的文档导航,查找,修改文档的方式.Beautiful Soup会帮你节省数小时甚至数天的工作时间.

> pip install requests -i https://pypi.tuna.tsinghua.edu.cn/simple


```python
from bs4 import BeautifulSoup
import requests
```

```python
url="http://www.zuihaodaxue.cn/zuihaodaxuepaiming2019.html"
```

```python
# 获取原始网页数据
def getHTMLText(url):
    try:
        r=requests.get(url,timeout=30)
        r.raise_for_status() #如果发送了一个错误请求(一个 4XX 客户端错误，或者 5XX 服务器错误响应)，我们可以通过 Response.raise_for_status() 来抛出异常
        r.encoding = r.apparent_encoding #防止乱码，从内容中分析出的响应内容编码方式。
        return r.text
    except:
        return "爬取失败"   
```

```python
# 测试下网页获取函数 
html=getHTMLText(url)
print(html)
```

```python
soup=BeautifulSoup(html)
print(soup.head)
```

```python
allUniv=[]
```

```python
def fillUinvList(soup):
    data = soup.find_all('tr') # 找到tr标签 
    #print(data)
    for tr in data:
        ltd = tr.find_all('td') # 找每个tr里的td标签
        #print(ltd)
        if len(ltd) == 0:
            continue
        
        singleUniv=[]
        for td in ltd:
            #print(type(td))
            singleUniv.append(td.string)
            #print(td.string)
        
        allUniv.append(singleUniv)
    
    
```

```python
fillUinvList(soup)
allUniv
```

```python
# 打印函数 
# chr(12288)  中文空格
def printUnivList(num):
    print("{1:^2}\t{2:{0}^10}\t{3:{0}^6}\t{4:{0}<6}".format(chr(12288),"排名","学校","省市","总分"))
    for i in range(num):
        u=allUniv[i]
        print("{1:^2}\t{2:{0}^10}\t{3:{0}^6}\t{4:{0}<6}".format(chr(12288),u[0],u[1],u[2],u[3]))

printUnivList(10)
```

```python
if __name__=="__main__":
    url="http://www.zuihaodaxue.cn/zuihaodaxuepaiming2019.html"
    html=getHTMLText(url)
    soup=BeautifulSoup(html)
    fillUinvList(soup)
    printUnivList(10)
    
```
